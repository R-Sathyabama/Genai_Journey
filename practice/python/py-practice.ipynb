python file handling

```python
## use read file handling
a=open("test.md",'r')
b=a.read()
print(b)
```

```python
##use write file handling
a=open("test.md",'w')
b=a.write("\ntesting writing file throught python \nMy first test")
print(b)
a.close()
```

```python
##use append file handling
#here instead of like this "a=open("test.md",'a')" we called file using 'with', 
# better to use 'with' format mostly in file handling
with open("test.md",'a') as a:                
    b=a.write("\n first file appending test")
    #we are trying append but usin append it will not support so instead 'append'  use 'write'
    print(b)   
    #here we  used 'with' so file is automatically closed here 'a. close()'  not needed.                           
```
